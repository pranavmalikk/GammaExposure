```python
import pandas as pd
```

```python
census_df = pd.read_csv('census.csv')
census_df.head()
```

```python
states = set(census_df.STNAME)
dict_of_states = {}

for state in states:
    counties_df = census_df[census_df.STNAME == state] #filter down to the specific state
    population = sum(counties_df.sort_values(by = 'CENSUS2010POP', ascending = False).CENSUS2010POP[1:4]) #within that state sum the three most populous counties
    dict_of_states[state] = population

df = pd.DataFrame.from_dict(dict_of_states, orient = 'index', columns = ['Populations'])
df = df.sort_values(by = 'Populations', ascending = False)
list(df.iloc[0:3].index)
```

```python
counties = census_df.groupby('STNAME').count()
x = counties.sort_values(by = 'CTYNAME', ascending = False)
x.index[0]
```

```python
max_pop_county1 = []
min_pop_county = []
largest_change = []
only_counties = census_df[census_df['SUMLEV'] == 50]
only_counties[only_counties.columns[9:15]]    
max_pop_county1 = only_counties[['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']].max(axis=1)
min_pop_county = only_counties[['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']].min(axis=1)
```

```python
census_df.loc[:, 'MAXPOP'] = max_pop_county1
census_df.loc[:, 'MINPOP'] = min_pop_county
largest_change = census_df["MAXPOP"] - census_df["MINPOP"]
census_df.loc[:, 'ABS_CHANGE'] = largest_change
x = census_df.sort_values(by = 'ABS_CHANGE', ascending = False)
x.iloc[0].CTYNAME
```
